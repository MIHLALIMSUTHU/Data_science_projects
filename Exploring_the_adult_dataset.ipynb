Replace salary values ['<=50k', '>50k'] With 0 and 1

```python
adult['income'].unique()
```

```python
adult['income'].value_counts()
```

```python
sns.countplot(x='income', data=adult, hue='income')
```

```python
def income_data (inc) :
  if inc == '<=50K' :
    return 0
  else :
    return 1
```

```python
adult['encoded_income']= adult['income'].apply(income_data)
```

```python
adult.head(1)
```

Which workclass is getting the highest salary?

```python
adult.columns
```

```python
highest_income_class=adult.groupby('workclass')['encoded_income'].mean().sort_values(ascending=False)
print(highest_income_class)
```

Who has better chances to get an income of >50k, male or female?

```python
adult.groupby('gender')['encoded_income'].mean().sort_values(ascending= False)
```
